```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("StudentsPerformance.csv")
df
```

A pie chart of showing male and female students with percentage

```python
labels = list(df['gender'].value_counts().keys())
values = list(df['gender'].value_counts().values)

plt.pie(values, labels=labels, wedgeprops={'ec': 'k'}, shadow=True, autopct='%1.2f%%')
plt.title('Male-Female Ratio')
plt.show()
```

A scatter plot which helps to analyse the relationship between the 'reading marks' and 'writing marks'

```python
# This plot shows us that the students with the higher reading score also har higher score in writing. 
# From the graph we can infer that there is a positive realtionship between 'Readinf Score' and 'Writing Score
```

```python
sns.relplot(x='reading score', y='writing score', data=df)
plt.show()
```

```python
# This plot shows the realtionship between 'Reading Score' and 'Writing Score' and
# also provides the information about the gender of the student
```

```python
sns.relplot(x='reading score', y='writing score', data=df, hue='gender')
plt.show()
```

```python
# This plot increases our scope of analysis by also encompassing the 'Race/Ethnicity' group of the students
```

```python
sns.relplot(x='reading score', y='writing score', data=df, hue='gender', style='race/ethnicity')
plt.show()
```

A bar chart to show the level of parental education for the students

```python
df['parental level of education'].value_counts()
```

```python
label = list(df['parental level of education'].value_counts().keys())
values = list(df['parental level of education'].value_counts().values)

plt.bar(label, values, color='green')
plt.xticks(rotation = 60)
plt.title('Parental Education Level')
plt.ylabel('No of Parents')
plt.xlabel('Education Level')
plt.show()
```

A pie chart to show the Educational Level of parents

```python
# This pie chart shows the Education Level of parents of the students along with the percantage of parents belonging
#  to each kind of Course or degree. It shows that Parents of only 5.9% students have a Master's Degree. 
# Parents of maximum students have either an Associates Degree (22.2%) or simply have attended some college (22.6%)
```

```python
label = list(df['parental level of education'].value_counts().keys())
values = list(df['parental level of education'].value_counts().values)

plt.pie(values, labels=label, wedgeprops={'ec':'k'}, shadow=True, autopct='%1.2f%%')
plt.title('Parental Education Pie Chart')
plt.show()
```

A Bar chart to show the preparation of test by each gender

```python
# We use countplot() function that creates a bar graph for a 
# catogical value by keeping the categories at x-axis and their count at y axis
```

```python
sns.countplot(x="test preparation course", hue="gender", data=df)
plt.title('Test Preparation (Gender wise)')
plt.xlabel('Test Preparation Course')
plt.ylabel('Count')
plt.show()
```

A Bar chart to show the lunch option chosen by each gender

```python
# From this graph, it can be infered that more than 300 Male and Female students have opted for 'Standard' lunch
```

```python
sns.countplot(x="lunch", hue="gender", data=df)
plt.title('Lunch option (Gender wise)')
plt.xlabel('Lunch Options')
plt.ylabel('Count')
plt.show()
```

Bar chart for showing Ethnicity with gender

```python
# From this graph, it can be infered that:

# Maximum male students are from Ethnic Group C
# Minimum male students are from Ethnic Group A
# Maximum female students are from Ethnic Group C
# Minimum female students are from Ethnic Group A
```

```python
sns.countplot(x='race/ethnicity', data=df, hue='gender')
plt.xlabel('Race / Ethnic Group')
plt.ylabel('Count')
plt.title('Ethnicity (Gender wise)')
plt.show()
```
